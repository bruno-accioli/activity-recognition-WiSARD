```python
import pickle

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
with open('results/model1_gs_results.pickle', 'rb') as file:
    gs_model1 = pickle.load(file)
    
with open('results/model2_gs_results.pickle', 'rb') as file:
    gs_model2 = pickle.load(file)
```

```python
gs_model1 = gs_model1.sort_values(['param_thermometer__n_bits', 'param_wisard__address_size']).reset_index()
gs_model2 = gs_model2.sort_values(['param_thermometer__n_bits', 'param_wisard__address_size']).reset_index()
```

Calculating the percentage of times model 2 got better results than model 1.

```python
(gs_model2['mean_test_score'] > gs_model1['mean_test_score']).mean()
```

```python
accuracy_increase = ((gs_model2['mean_test_score'] - gs_model1['mean_test_score']) /gs_model1['mean_test_score']) * 100
ax = sns.lineplot(x=gs_model1['param_thermometer__n_bits'], y=accuracy_increase)
ax.set_title('Aumento percentual da acurácia média', size=14)
ax.set_ylabel('Aumento de acurácia (%)', size=12)
ax.set_xlabel('Número de bits do termômetro', size=12)
ax.set_xticks([40, 50, 60, 70, 80, 90])
ax.set_ylim(0, 6)
sns.despine()
```

```python
accuracy_increase = ((gs_model2['mean_test_score'] - gs_model1['mean_test_score']) /gs_model1['mean_test_score']) * 100
ax = sns.lineplot(x=gs_model1['param_wisard__address_size'], y=accuracy_increase)
ax.set_title('Aumento percentual da acurácia média', size=14)
ax.set_ylabel('Aumento de acurácia (%)', size=12)
ax.set_xlabel('Tamanho do endereço (WiSARD)', size=12)
ax.set_ylim(0, 7)
sns.despine()
```
